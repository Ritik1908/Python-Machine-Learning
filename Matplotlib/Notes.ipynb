```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

Pie charts are generally used to show percentage or proportional data and usually the percentage represented by each category is provided next to the corresponding slice of pie.

```python
areas = ['marketing', 'Sales', 'Development', 'HR', 'Customer Support']
budget = [2.5, 4, 10, 1, 6]
```

```python
plt.pie(budget, labels=areas)
```

To remove technicailical details shown in above plot we need a plt.show() function, this only prints the graph.

```python
plt.pie(budget, labels=areas)
plt.show()
```


The above graph doesn't show %age of data distribution, it's hard to visualize.
If we want to see the percentage of distribution, there is a parameter "autopct" in pie function that we can set to get the percentage of distribution in out data.

```python
plt.pie(budget, labels=areas, autopct="%0.2f%%")
plt.show()
```

# Scatter Plot

A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

```python
x = [23, 43, 52, 3, 4, 42, 4, 24]
y = [43, 41, 12, 4, 5, 32, 6, 62]
plt.scatter(x, y)
```

# Annotations we can use in matplotlib library

The following functions are most commonly used in matplotlib to show annotations in graphs to increase it's readability.
    1. xlabel() -> Annotations of values in x-axis.
    2. ylabel() -> Annotation of values in y-axis.
    3. title() -> To show the title of the graph obtained.

```python
x = [23, 43, 52, 3, 4, 42, 4, 24]
y = [43, 41, 12, 4, 5, 32, 6, 62]
plt.scatter(x, y)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary Graph")
plt.show()
```

# 2D plot

In matplotlib we have, plt.plot() function to draw 2nd plots.

```python
product_id = [1, 2, 3, 4, 5, 6, 7, 8]
online_price = [233, 456, 770, 120, 222, 444, 200, 300]
offline_price = [400, 300, 100, 245, 423, 342, 120, 130]
chor_bazaar_price = [100, 200, 200, 150, 230, 100, 200, 120]
```

```python
plt.plot(product_id, online_price)
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.show()
```

### We can change color of graph in matplotlib library with "color" parameter in function.
color takes commonly used color names.

```python
plt.plot(product_id, online_price, color='red')
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.show()
```

### We have "marker" parameter to set the marker in graph
Commonly used values:
    1. "+" -> Plus marker
    1. "D" -> Diamond

```python
plt.plot(product_id, online_price, color='red', marker='D')
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.show()
```

```python
# To increase marker size
plt.plot(product_id, online_price, color='red', marker='D', markersize=10)
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.show()
```

## Comparitive Study Between Different attributed using 2D plots

It will plot the entire thing in same graph.

```python
plt.plot(product_id, online_price, color='red', marker='D', markersize=10)
plt.plot(product_id, offline_price, color='yellow', markersize=10)
plt.plot(product_id, chor_bazaar_price, color='green', marker='+', markersize=10)
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.show()
```

# Legend In Graph
To assign a legend to our plots, we need to give each plot a "label" attribute.
After giving "label" to each plot, we need to call
    "plt.legend()"
To get out legend printed in plot. (It takes free space available on plot to print itself.)

```python
plt.plot(product_id, online_price, color='red', marker='D', markersize=10, label="Online Prices")
plt.plot(product_id, offline_price, color='yellow', markersize=10, label='Offline Prices')
plt.plot(product_id, chor_bazaar_price, color='green', marker='+', markersize=10, label='Chor Bazaar Prices')
plt.xlabel("Product")
plt.ylabel("Prices")
plt.title("Product Price Comparision")
plt.legend()
plt.show()
```

# Bar Graph 

```python
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
price = [65000, 70000, 75000, 50000, 52000, 80000, 85000]
```

```python
plt.bar(year, price)
plt.xlabel('Year')
plt.ylabel('Price (in INR)')
plt.title("Year by Year Price Comparision")
plt.show()
```

### Changing color of bargraph

```python
plt.bar(year, price, color='red')
plt.xlabel('Year')
plt.ylabel('Price (in INR)')
plt.title("Year by Year Price Comparision")
plt.show()
```

### Changing axis of bargraph

```python
plt.barh(year, price, color='red')
plt.xlabel('Year')
plt.ylabel('Price (in INR)')
plt.title("Year by Year Price Comparision")
plt.show()
```
