```python
import seaborn as sns
import matplotlib as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('titanic')
```

```python
data.head()
```

```python
# Problem 1 : Plot a Heatmap on the correlation between multiple columns
x = data.corr() # Finding correlation between different columns (kind of pivot table format)
x
```

```python
sns.heatmap(x, linewidth=1)
```

```python
sns.heatmap(x, linewidth=1, annot=True, fmt='.1f') # Heatmap with correlation values upto 1 decimal place.
```

```python
# Problem 2 : Plot the distribution of age column using a Boxplot on the basis of sex column
sns.catplot(x="sex", y="age", kind="box", data=data)
```

```python
# Problem 3 : Plot the distribution of age using a ViolinPlot on the basis of both Pclass and Sex column
sns.catplot(x="pclass", y="age", hue="sex", split=True, kind="violin", data=data)
```

```python
# Problem 4 : Plot the count of Survivors and Non-Survivors
sns.catplot(x='survived', kind='count', data=data)
```

```python
# Problem 5: Plot the count of survivors as well as non-survivors on the basis of Pclass
sns.catplot(x='survived', hue='pclass', kind='count', data=data)
```

```python
# Problem 6: Plot a scatterplot between age and fare column on the basis of both Pclass and Sex column
```

```python
# Plot 1:
sns.relplot(x='age', y='fare', hue='pclass', size='sex', sizes=(20, 200), data=data)
```

```python
# Plot 2
sns.relplot(x='age', y='fare', hue='pclass', style='sex', data=data)
```
