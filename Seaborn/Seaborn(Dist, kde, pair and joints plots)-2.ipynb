```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
# Problem 1: Plot a rugplot on given below random generated data.
num = np.random.randn(170)
sns.distplot(num, hist=False, kde=False, rug=True)
```

```python
# Problem 2: Plot a pairs plot on iris data set showing different species. Write your conclusions as well.
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris, hue='species')
```

# Conclusions:
    1. petal_length can be used to seperate 'setosa' from others
    2. There is linear relationship between all features.
    3. 'setosa' have low variation in it's length and width features
    4. 'versicolor' and 'verginica' have almost identical length and width features.

```python
# Problem 3: Plot a joint plot showing relation between 'petal length' and 'petal width' with density estimates.
sns.jointplot(x='petal_length', y='petal_width', kind='kde', data=iris)
```

```python
# Problem 4: Plot a pie chart of survived based on sex.
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic = titanic[titanic['survived'] == 1]
data = titanic.groupby('sex')['sex'].count()
label = data.index
values = data.values
```

```python
plt.pie(values, labels=label, autopct="%0.1f%%")
plt.title("Distribution of gender who survived")
plt.show()
```
